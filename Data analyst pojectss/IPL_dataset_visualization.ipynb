```python

match_win=df['match_winner'].value_counts()
sns.barplot(y=match_win.index, x=match_win.values,palette='rainbow')
plt.title('Most match win by a team')
plt.xlabel('Match win by a team')
plt.ylabel('Teams name')
```

```python
sns.countplot(x=df['toss_decision'],palette='viridis')
plt.title('Toss decision trends')
```

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage = (count * 100)/df.shape[0]
percentage.round(2)
```

```python
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values, y=count.index,palette='viridis', orient='h',legend=False)
plt.title('Top 10 Players with Most Man of the Match')
plt.xlabel('Match Awards')
plt.ylabel('Players Name')
plt.xticks(range(0, int(max(count.values)) + 1))
plt.show()
```

```python
high_scorer=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
high_scorer
```

```python
high_scorer.plot(kind='barh')
```

```python
df['highest_wicket']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wicket']=df['highest_wicket'].astype(int)
Top_bowler=df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
Top_bowler
```

```python
sns.barplot(x=Top_bowler.values,y=Top_bowler.index,palette='Spectral',hue=Top_bowler.index,legend=False)
plt.title("Top Bowlers")
plt.xlabel("Wickets")
plt.ylabel("Bowler Name")
plt.show()
```
